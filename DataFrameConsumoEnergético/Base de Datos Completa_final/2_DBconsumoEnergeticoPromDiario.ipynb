```python
import pandas as pd
import numpy as np
from datetime import datetime, time
```

```python
from pandas import read_excel

df_hora=pd.read_excel('df_CE_prom_hora_terminado.xlsx', index_col=False)
```

```python
df_hora['Timestamp'] = pd.to_datetime(df_hora['Timestamp'])
df_hora = df_hora.groupby([df_hora['Timestamp'].dt.date, df_hora['Timestamp'].dt.day_name()]).mean()
```
