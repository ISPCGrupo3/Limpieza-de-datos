```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
dfconsumo=pd.read_excel('tablasConcatenadas.xlsx')
#dfconsumo.tail()
```

```python
null_columns = dfconsumo.columns[dfconsumo.isnull().any()]
dfconsumo[null_columns].isnull().sum()
print(dfconsumo[dfconsumo.isnull().any(axis=1)][null_columns].head())
```

```python
for col in list(dfconsumo.columns)[1:]:
    for i in range(len(dfconsumo[col])):
        if np.isnan(dfconsumo[col][i]):
            if i == 0:
                dfconsumo[col][i] = dfconsumo[col][i+1]
            elif i == len(dfconsumo[col]) - 1:
                dfconsumo[col][i] = dfconsumo[col][i-1]
            else:
                dfconsumo[col][i] = (dfconsumo[col][i+1] + dfconsumo[col][i-1])/2

dfconsumo
                
```

```python
#no se completan los datos cuando hay varias filas vacías no puede calcular sobre ceros
dfconsumo.to_excel('2104_lleno.xlsx', index=False)
```

```python
#enlas celdas de las filas vacías de corrido completa con el promedio de la columna
dfconsumo = dfconsumo.fillna(dfconsumo.mean())
```

```python

dfconsumo.to_excel('2104_prom.xlsx', index=False)
```

```python
#comprobamos que no hay nulls
null_columns = dfconsumo.columns[dfconsumo.isnull().any()]
dfconsumo[null_columns].isnull().sum()
print(dfconsumo[dfconsumo.isnull().any(axis=1)][null_columns].head())
```

```python
#Agregamos columnas Suma TRs y atribuimos la diferencia con MT por consumo Cine

dfconsumo['Suma-TR-1,2,3,4-(Kw)']=dfconsumo.loc[:, [
    'COCS.COCS-TR-01\nReal Power\n(kW)', 'COCS.COCS-TR-02\nReal Power\n(kW)', 'COCS.COCS-TR-03\nReal Power\n(kW)', 'COCS.COCS-TR-04\nReal Power\n(kW)']].sum(axis=1)
dfconsumo['Diferencia_cine']=dfconsumo.loc[:, 'COCS.COCS-MT-01\nReal Power\n(kW)'] - dfconsumo.loc[:,'Suma-TR-1,2,3,4-(Kw)']
df_sumaTR_dif_cine=dfconsumo
df_sumaTR_dif_cine.head()
```

```python
#Promedios consumos con suma TRs y diferencia cine

dfProm_sumaTR_dif_cine = df_sumaTR_dif_cine.fillna(df_sumaTR_dif_cine.mean())
dfProm_sumaTR_dif_cine.head()
```

```python
dfProm_sumaTR_dif_cine.to_excel('Prom_Compl_2104.xlsx',index=False)
```

```python
df_sumaTR_dif_cine['Timestamp'] = pd.to_datetime(
    df_sumaTR_dif_cine['Timestamp'])
df_sumaTR_dif_cine = df_sumaTR_dif_cine.groupby(
    [df_sumaTR_dif_cine['Timestamp'].dt.date, df_sumaTR_dif_cine['Timestamp'].dt.hour]).mean()
```
