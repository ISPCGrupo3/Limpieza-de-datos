```python
import pandas as pd
import numpy as np
from datetime import datetime, time
```

```python
# Leemos el archivo 'tablasConcatenadas.xlsx'
dfconsumo=pd.read_excel('tablasConcatenadas.xlsx')
dfconsumo.tail()
```

```python
#Corroboramos la existencia de datos null
def control_nulls(df):
    null_columns = dfconsumo.columns[dfconsumo.isnull().any()]
    dfconsumo[null_columns].isnull().sum()
    print(dfconsumo[dfconsumo.isnull().any(axis=1)][null_columns].head())

control_nulls(dfconsumo)
```

```python
#Creamos un ciclo for para iterar columnas y 
# completar valores con el resultado de la media de el valor anterior y posterior de la columna correspondiente

for col in list(dfconsumo.columns)[1:]:
    for i in range(len(dfconsumo[col])):
        if pd.isna(dfconsumo[col][i]):
            if i == 0:
                dfconsumo[col][i] = dfconsumo[col][i+1]
            elif i == len(dfconsumo[col]) - 1:
                dfconsumo[col][i] = dfconsumo[col][i-1]
            else:
                dfconsumo[col][i] = (dfconsumo[col][i+1] + dfconsumo[col][i-1])/2

# Observamos las primeras 5 filas
dfconsumo.head()
                
```

```python
#Corroboramos la existencia de datos null

control_nulls(dfconsumo)
```

```python
# completamos con la media de los valores de cada columna los valores que aún continúan vacíos 
# ( porque no tienen valor anterior y posterior para calcular la media con el ciclo for)

dfconsumo = dfconsumo.fillna(dfconsumo.mean())
```

```python
#Corroboramos la existencia de datos null
control_nulls(dfconsumo)
```

```python
#Creamos un archivo excel 'dfConsumoEnergeticoCompleto.xlsx'
dfconsumo.to_excel('df_1_tablasCE_sin_nulls.xlsx', index=True)
```

```python
#Agregamos columnas Suma TRs y atribuimos la diferencia con MT por consumo Cine

dfconsumo['Suma-TR-1,2,3,4-(Kw)']=dfconsumo.loc[:, [
    'COCS.COCS-TR-01\nReal Power\n(kW)', 'COCS.COCS-TR-02\nReal Power\n(kW)', 'COCS.COCS-TR-03\nReal Power\n(kW)', 'COCS.COCS-TR-04\nReal Power\n(kW)']].sum(axis=1)
dfconsumo['Diferencia_cine']=dfconsumo.loc[:, 'COCS.COCS-MT-01\nReal Power\n(kW)'] - dfconsumo.loc[:,'Suma-TR-1,2,3,4-(Kw)']

# Observamos las primeras o las últimas filas con head o tail
dfconsumo.head(120)
```

```python

#Corroboramos la existencia de datos null
control_nulls(dfconsumo)
```

```python
# Separamos columna 'Timestamp' en Fecha y Hora con datetime y nombre del día
# calculamos promedios de consumo energético agrupando datos por hora

dfconsumo['Timestamp'] = pd.to_datetime(dfconsumo['Timestamp'],format='%d/%m/%Y %H:%M:%S')
dfconsumo = dfconsumo.groupby([dfconsumo['Timestamp'].dt.date,
                              dfconsumo['Timestamp'].dt.hour,dfconsumo['Timestamp'].dt.day_name()]).mean()
```

```python
# Observamos las primeras o últimas filas
dfconsumo.head(36)
```

```python
#Creamos un archivo excel 'df_2_CE_SumaTRs_DifCine.xlsx'
dfconsumo.to_excel('df_2_CE_SumaTRs_DifCine.xlsx', index=True)
```

```python
#Creamos un ciclo for para volver a combinar valores de la columna 'Timestamp'
data = []
for index, row in dfconsumo.iterrows():
    data.append([datetime.combine(index[0], time(hour=index[1])),index[2]] +
        list(row.values))
```

```python
dfconsumo = pd.DataFrame(data, columns=['Timestamp','Día']+ list(dfconsumo.columns))
```
